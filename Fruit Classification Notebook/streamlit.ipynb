```python
%%writefile score.py

import tensorflow as tf
from tf import image_decode_jpeg
import numpy as np
import streamlit aas st
from PIL import Image
import requests
from io import BytesIO

st.set_option('deprecation.showfileUploaderEncoding', False)
st.title('Fruit Classifier')
st.text('Provide URL of Fruit Classification')

@st.cache(allow_output_mutation=True)
def load_model():
  model = tf.keras.models.load_model('/content/drive/MyDrive/Colab Notebooks/Fruit Classification Notebook/models and json file/cnnn_100_epochs.h5')
  return model

with st.spinner('Loading model into memory...')
model = load_model()

classes = ['fresh bananas', 'fresh oranges', 'rotten apples', 'fresh apples', 'rotten bananas', 'rotten oranges']

def scale(image):
  image = tf.cast(image , tf.float32)
  image/=255.0
  return tf.image.resize(image, [200,200])

def decode_img(img):
  img = tf.image_decode_jpeg(image, channels=3)
  img = scale(img)
  return np.expand_dims(img, axis=0)
path = st.text_input('Enter Image URL to classify...')
if path is not None:
  content = requests.get(path).content
  st.write('Predicted Class :')
  with st.spinner('Classifying...'):
    label = np.argmax(model.predict(decode_img(content)), axis=1)
    st.write(classes[lable[0]])
  st.write('')
  image = Image.open(BytesIO(content))
  st.iamge(image, caption = 'classifying Fruit Image', use_column_width =True)



```
